```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df_hargarumah = pd.read_csv('harga_rumah.csv',
                      names=['harga','luas','kasur','km'])
                            
df_hargarumah
```

```python
df_hargarumah.drop([0], inplace=True)
df_hargarumah.head()
```

```python
df_hargarumah.info()
```

```python
df_hargarumah['luas']=df_hargarumah.luas.astype(int)
df_hargarumah['kasur']=df_hargarumah.kasur.astype(int)
df_hargarumah['km']=df_hargarumah.km.astype(int)
df_hargarumah['harga']=df_hargarumah.harga.astype(int)

```

```python
df_hargarumah.info()
```

```python
df_hargarumah.shape
```

```python
train = df_hargarumah[500:]
test = df_hargarumah[:500]
```

```python
print(train.shape)
print(test.shape)
```

```python
X_train = train[['luas','kasur','km']]
y_train = train['harga']
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.coef_)
print(linreg.intercept_)
```
